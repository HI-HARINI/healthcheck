```python
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
```

```python
#Task 1 
#Sort the data as per total number of cases
sorted_df = df.sort_values(by=['total_cases'])
sorted_df
```

```python
#Task 2
#Get top 5 countries who has the least number of cases and plot a bar graph
heavy_satellite_5 = sorted_df.head(5)
print(heavy_satellite_5)

name = heavy_satellite_5['country']
weight = heavy_satellite_5['total_cases']

plt.xlabel("country")
plt.xticks(rotation='vertical')
plt.ylabel("total_cases")


label = name
value = weight 
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue')) #bar-grap
```

```python
#Task 3
#Sort the data as per total number of deaths
sorted_df = df.sort_values(by=['total_deaths'])
sorted_df
```

```python
#Task 4
#Get top 5 countries who has the maximum number of deaths and plot a bar graph
heavy_satellite_5 = sorted_df.tail(5)
print(heavy_satellite_5)

name = heavy_satellite_5['country']
weight = heavy_satellite_5['total_deaths']

plt.xlabel("country")
plt.xticks(rotation='vertical')
plt.ylabel("total_deaths")


label = name
value = weight 
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue')) #bar-grap

```

```python
#Task 5
#Sort the data as per active cases
sorted_df = df.sort_values(by=['active_cases'])
sorted_df
```

```python
#Task 6
#Get top 5 countries who has the maximum number of active cases and plot a bar graph
heavy_satellite_5 = sorted_df.tail(5)
print(heavy_satellite_5)

name = heavy_satellite_5['country']
weight = heavy_satellite_5['active_cases']

plt.xlabel("country")
plt.xticks(rotation='vertical')
plt.ylabel("active_cases")


label = name
value = weight 
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue')) #bar-grap
```
